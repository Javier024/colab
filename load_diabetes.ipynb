```python
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.svm import SVC
from sklearn.pipeline import make_pipeline
from sklearn.metrics import classification_report
import numpy as np

# Cargar el dataset
diabetes = load_diabetes()
X = diabetes.data
y = diabetes.target

# Convertir a clasificaciÃ³n binaria
y_class = np.where(y > np.median(y), 1, 0)

# DivisiÃ³n entrenamiento / prueba
X_train, X_test, y_train, y_test = train_test_split(X, y_class, test_size=0.2, random_state=42)

# Crear pipeline con PCA + SVM
pca = PCA(n_components=5, whiten=True, random_state=42)  # solo 5 componentes por tener pocas features
svc = SVC(kernel='rbf', class_weight='balanced', random_state=42)
model = make_pipeline(pca, svc)

# Entrenar
model.fit(X_train, y_train)

# Predecir
y_pred = model.predict(X_test)

# Evaluar
print(classification_report(y_test, y_pred, target_names=['bajo', 'alto']))
```

informe con precisiÃ³n, recall y F1-score comparando el rendimiento del modelo para clasificar pacientes como con nivel "bajo" o "alto" de enfermedad.

Por Ãºltimo, podemos utilizar la validaciÃ³n cruzada de bÃºsqueda en cuadrÃ­cula para explorar combinaciones de parÃ¡metros. AquÃ­ ajustaremos C (que controla la dureza del margen) y gamma (que controla el tamaÃ±o del nÃºcleo de la funciÃ³n de base radial), y determinaremos el mejor modelo:

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report
import numpy as np

# Cargar y preparar los datos
X, y = load_diabetes(return_X_y=True)
y = np.where(y > np.median(y), 1, 0)  # ClasificaciÃ³n binaria (mayor o menor que la mediana)

# DivisiÃ³n entrenamiento / prueba
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.2, random_state=12)

# Entrenar el modelo SVM con kernel lineal
model = SVC(kernel='linear', C=4)
model.fit(Xtrain, ytrain)
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV

param_grid = {'C': [1, 5, 10, 50], 'gamma': [0.0001, 0.0005, 0.001, 0.005]}

# Realizar GridSearchCV
grid = GridSearchCV(model, param_grid, cv=5)
%time grid.fit(Xtrain, ytrain)

# Imprimir los mejores parÃ¡metros encontrados
print(grid.best_params_)
```

```python
model = grid.best_estimator_
yfit = model.predict(Xtest)
```

ytest: Son las etiquetas reales del conjunto de prueba.

Salida Esperada:
El reporte de clasificaciÃ³n incluirÃ¡ las siguientes mÃ©tricas para cada clase (0 = bajo, 1 = alto):

Precision: CuÃ¡nto de lo predicho como positivo es realmente positivo.

Recall: CuÃ¡nto de lo que es realmente positivo fue identificado como positivo.

F1-Score: La media armÃ³nica entre precisiÃ³n y recall.

Support: El nÃºmero de instancias en cada clase.

```python
from sklearn.metrics import classification_report

# Imprimir el reporte de clasificaciÃ³n para el dataset de diabetes
print(classification_report(ytest, yfit, target_names=['bajo', 'alto']))
```

```python
from sklearn.metrics import confusion_matrix

# Generar la matriz de confusiÃ³n
mat = confusion_matrix(ytest, yfit)

# Visualizarla con un mapa de calor
plt.figure(figsize=(8, 6))
sns.heatmap(mat.T, square=True, annot=True, fmt='d',
            cbar=False, cmap='Blues',
            xticklabels=['bajo', 'alto'],
            yticklabels=['bajo', 'alto'])

# Etiquetas y tÃ­tulo
plt.xlabel('Etiqueta Real')
plt.ylabel('Etiqueta Predicha')
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

# Cargar los datos de diabetes
from sklearn.datasets import load_diabetes
X, y = load_diabetes(return_X_y=True)

# Convertir el problema a binario (mayor que la mediana)
y = (y > np.median(y)).astype(int)  # 1 = alto, 0 = bajo

# Dividir los datos en conjunto de entrenamiento y prueba
Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.2, random_state=12)

# Entrenar el modelo SVM
svm_model = SVC(kernel='linear', C=4)
svm_model.fit(Xtrain, ytrain)
svm_predictions = svm_model.predict(Xtest)

# Entrenar el modelo Ãrbol de DecisiÃ³n
tree_model = DecisionTreeClassifier(random_state=12)
tree_model.fit(Xtrain, ytrain)
tree_predictions = tree_model.predict(Xtest)

# Imprimir el reporte de clasificaciÃ³n para SVM
print("SVM Classification Report:")
print(classification_report(ytest, svm_predictions, target_names=['bajo', 'alto']))

# Imprimir el reporte de clasificaciÃ³n para Ãrbol de DecisiÃ³n
print("\nÃrbol de DecisiÃ³n Classification Report:")
print(classification_report(ytest, tree_predictions, target_names=['bajo', 'alto']))
```
